```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

## 条件

```python
v0 = 20.0 # 初速度 [m/s]
g  = 9.8 # 重力加速度 [m/s2]
t_range = 4 # 計算する時間範囲 [s]
```

## 解析解

```python
t = np.linspace(0, t_range, 100)
z_an = v0*t - 1.0/2.0*g*t*t
```

## 数値解

```python
dt = 0.1 # 微小時間 [s]
n = int(t_range/dt) # 計算ステップ数 計算時間範囲を微小時間で割ったもの
t_nu = [0] # 各計算ステップの時刻を保持するリスト
z_nu = [0] # 各計算ステップの位置を保持するリスト
for n in range(n):
    t_next = t_nu[-1] + dt # 次の時刻を算出
    z_next = z_nu[-1] + (v0 - g*t_nu[-1])*dt # 次の時刻の位置を算出
    t_nu.append(t_next)
    z_nu.append(z_next)
```

## グラフで比較

```python
plt.plot(t, z_an, label="analytical")
plt.plot(t_nu, z_nu, label="numerical dt = 0.1")
plt.xlabel("t [s]")
plt.ylabel("z [m]")
plt.legend()
```
